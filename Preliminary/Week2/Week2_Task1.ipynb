```python
import pandas as pd
import numpy as np

ema_df = pd.read_csv('../../Dataset/EMA/general_ema.csv')
sensing_df = pd.read_csv('../../Dataset/Sensing/sensing.csv')
```

```python

# Define columns we want for each EP
ep_column_names = [col for col in sensing_df.columns if '_ep_' in col]
ep_column_prefixes = set()
for col in ep_column_names:
    # Split on '_ep_' and take the first part
    prefix = col.split('_ep_')[0]
    ep_column_prefixes.add(prefix)

print(ep_column_prefixes)
```

```python
def create_ep_profile(prefix_list):
    # Initialize empty list to store data
    # Create new dataframe with repeated rows for each EP
    ep_profile = pd.DataFrame({
        'uid': sensing_df['uid'].repeat(3),
        'date': sensing_df['day'].repeat(3),
        'ep': np.tile([1, 2, 3], len(sensing_df))
    })
    # Add each metric for this EP
    for prefix in prefix_list:
    # Get values for all three EPs at once
        ep_profile[prefix] = pd.concat([
            sensing_df[f"{prefix}_ep_1"],
            sensing_df[f"{prefix}_ep_2"],
            sensing_df[f"{prefix}_ep_3"]
        ]).reset_index(drop=True)
    
    ep_profile = ep_profile.set_index(['uid', 'date', 'ep'])
    
    return ep_profile
```

```python
ep_profile = create_ep_profile(ep_column_prefixes)
display(ep_profile)
```
