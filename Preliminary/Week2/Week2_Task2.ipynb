```python
import pandas as pd
import numpy as np

ema_df = pd.read_csv('../../Dataset/EMA/general_ema.csv')
sensing_df = pd.read_csv('../../Dataset/Sensing/sensing.csv')
```

```python

# Convert day to datetime to get day of week
sensing_df['date'] = pd.to_datetime(sensing_df['day'].astype(str), format='%Y%m%d')
sensing_df['day_of_week'] = sensing_df['date'].dt.strftime('%a')

display(sensing_df.head(15))
```

```python
# Get all full day columns
full_day_columns = [col for col in sensing_df.columns if col.endswith('_ep_0')]
print(full_day_columns)
full_day_profile = sensing_df[['uid', 'day', 'day_of_week'] + full_day_columns].copy()
```

```python
full_day_profile = full_day_profile.set_index(['uid', 'day'])
```
