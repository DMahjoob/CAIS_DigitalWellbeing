```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ema_df = pd.read_csv('../../Dataset/EMA/general_ema.csv')
sensing_df = pd.read_csv('../../Dataset/Sensing/sensing.csv')

# Get users with social_level responses (non-null values)
users_with_social = ema_df[ema_df['social_level'].notna()]
# Retain only the columns needed
users_with_social = users_with_social[['uid', 'day', 'social_level']]

# Calculate percentage of students with social_level responses
total_students = len(ema_df['uid'].unique())
students_with_social = len(users_with_social['uid'].unique())
percentage = (students_with_social / total_students) * 100

print(f"Percentage of students with social_level responses: {percentage:.2f}%")
print(f"({students_with_social} out of {total_students} students)")
```

```python
merged_df = users_with_social.merge(
    sensing_df,
    on=['uid', 'day'],
    how='left'
)
display(merged_df)
```
