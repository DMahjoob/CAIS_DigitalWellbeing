```python
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("sensing.csv")
demo = pd.read_csv("demographics.csv")
data.head(10)
```

```python
print(data.shape)
print(data.columns.tolist())
```

```python
# Merge demographics information
data = data.merge(demo[['uid', 'gender', 'race']], on='uid', how='left')

# Break down by EP
unlock_durations = data[['unlock_duration_ep_0', 'unlock_duration_ep_1', 'unlock_duration_ep_2', 'unlock_duration_ep_3']].sum()
print("Unlock Durations by EP:\n", unlock_durations)
```

```python
# Create a bar chart of the duration of unlocks in each EP across all students
plt.figure(figsize=(12, 6))
unlock_durations.plot(kind='bar')
plt.title('Duration of Unlocks by EP (All Students)')
plt.xlabel('EP')
plt.ylabel('Duration of Unlocks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# compare gender (each bar of the above plot will become 2 bars)
unlock_duration_by_day_gender = data.groupby(['gender'])[['unlock_duration_ep_0', 'unlock_duration_ep_1', 'unlock_duration_ep_2', 'unlock_duration_ep_3']].sum()
display(unlock_duration_by_day_gender)

# compare race (each bar will become X where X is the number of races)
unlock_duration_by_day_race = data.groupby(['race'])[['unlock_duration_ep_0', 'unlock_duration_ep_1', 'unlock_duration_ep_2', 'unlock_duration_ep_3']].sum()
display(unlock_duration_by_day_gender)
```
