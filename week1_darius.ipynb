```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
demographics_df = pd.read_csv('sample_data/demographics.csv')

# Demographics breakdown with percentages
demographics_breakdown = demographics_df.groupby(['race', 'gender']).size().reset_index(name='count')
demographics_breakdown['percentage'] = (demographics_breakdown['count'] / demographics_df.shape[0]) * 100
print(demographics_breakdown)

# Plotting
demographics_pivot = demographics_breakdown.pivot(index='race', columns='gender', values='count').fillna(0)
demographics_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))

# Customize plot
plt.title('Distribution of Demographics by Race and Gender')
plt.xlabel('Race')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.tight_layout()

plt.show()
```

```python
# Read in Sensing Data
sensing_df = pd.read_csv('sample_data/sensing.csv')

locations = ["study", "social", "self_dorm", "other_dorm", "home", "food"]
unlock_num_cols = [f"loc_{location}_unlock_num" for location in locations]

# Rank by num unlocks
total_unlock_df = pd.DataFrame(sensing_df[unlock_num_cols].sum(axis=1), columns=['total_unlock_num'])
total_unlock_df = total_unlock_df.sort_values(by='total_unlock_num', ascending=False)
total_unlock_df.index.name = 'user_id'
total_unlock_df

plt.figure(figsize=(10, 6))
plt.bar(total_unlock_df.index, total_unlock_df['total_unlock_num'], color='skyblue')

# Adding labels and title
plt.title('Total Unlock Counts per User')
plt.xlabel('User ID')
plt.ylabel('Total Unlock Count')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability

# Display plot
plt.tight_layout()
plt.show()
```

```python
unlock_duration_cols = [f"loc_{location}_unlock_duration" for location in locations]

# Rank by total screentime
total_unlock_duration_df = pd.DataFrame(sensing_df[unlock_duration_cols].sum(axis=1), columns=['total_unlock_duration'])
total_unlock_duration_df = total_unlock_duration_df.sort_values(by='total_unlock_duration', ascending=False)
total_unlock_duration_df.index.name = 'user_id'
total_unlock_duration_df

# Display plot
plt.figure(figsize=(10, 6))
plt.bar(total_unlock_duration_df.index, total_unlock_duration_df['total_unlock_duration'], color='blue')
plt.title('Total Unlock Duration by User')
plt.xlabel('User ID')
plt.ylabel('Total Unlock Duration (seconds)')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
# Rank by sleep duration
sleeping_duration_df = sensing_df[["sleep_duration"]].sort_values(by='sleep_duration', ascending=False)
sleeping_duration_df.index.name = 'user_id'
sleeping_duration_df

# Display plot
plt.figure(figsize=(10, 6))
plt.bar(sleeping_duration_df.index, sleeping_duration_df['sleep_duration'], color='green')
plt.title('Total Sleep Duration by User')
plt.xlabel('User ID')
plt.ylabel('Total Sleep Duration (seconds)')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
# Rank by sleep time
sleeping_start_df = sensing_df[["sleep_start"]].sort_values(by='sleep_start', ascending=False)
sleeping_start_df.index.name = 'user_id'
sleeping_start_df

plt.figure(figsize=(10, 6))
plt.bar(sleeping_start_df.index, sleeping_start_df['sleep_start'], color='blue')
plt.title('Initial Sleep by User')
plt.xlabel('User ID')
plt.ylabel('Initial Sleep Time (seconds)')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
